```python
import struct
```

```python
buffer = bytearray()
print(buffer)
```

```python
start_ = 0xAA.to_bytes()
stop_ = 0xBB.to_bytes()

print(start_, stop_)
print(type(start_), type(stop_))
```

```python
buffer += start_
print(buffer)
```

```python
if buffer.startswith(start_):
    print("Head check ok")
```

```python
vals = [1.23, 3.454, 5.454]
for val in vals:
    buffer.extend(struct.pack('f', val))

buffer += stop_
```

```python
print(buffer)
```

```python
if buffer.startswith(start_) and buffer.endswith(stop_):
    print("struct is correct")
```

```python
print(f"{len(buffer)} bytes")

print(f"size of start and stop bytes: {len(start_) + len(stop_)}")
```
