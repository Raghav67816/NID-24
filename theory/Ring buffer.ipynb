### Understanding ring-buffer/circular buffer

Basic idea:

Consider an array [1, 2, 3, 4, 5]
incoming data is [6, 7, 8]

the array should become [1, 2,3 4, 5, 6, 7, 8]
for NID-24 latest data should be last

```python
import numpy as np
```

```python
buffer = np.linspace(0, 30, num=50)
incoming_data = np.linspace(0, 20, num=10)

print(buffer[-9])
```

```python
# Now keep appending data
index = 0
for val in incoming_data:
    if index >= len(buffer):
        index = 0
    buffer[index] = val
    index += 1
        
# now the latest should be first
print(buffer)
```

### these value should have been appended to the last not the first

```python
print(buffer)
index = 0
for val in incoming_data:
    if index >= len(buffer):
        index = 0
    buffer[:-1] = buffer[1:]
    buffer[index]
    index += 1
print(buffer)
```
