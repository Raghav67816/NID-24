### Simulating EMG signals

For now, we are not adding noise

```python
# import dependencies
import numpy as np
from time import sleep
import matplotlib.pyplot as plt
```

```python
# define parameters
FREQ = 15000 # hz
PHASE_DIFF = 0
STEPS = 100 # seconds
```

```python
rng = np.random.default_rng()
points = np.array([])
def simulate():
    global points
    points = np.array([])
    for i in range(STEPS):
        amplitude = rng.uniform(0.0*10e-3, 10e-3)
        omega = 2 * np.pi * FREQ
        y = amplitude * np.sin(omega * i)  # phase = 0
        sleep(1 / FREQ)
        points = np.append(points, y)

print(points.shape)
```
