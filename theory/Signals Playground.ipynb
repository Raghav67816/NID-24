## Convert 8 Channel Data File to 3 Channel Data File

These files will then be loaded from **EMG Signals Simulator App**, so we get actual recorded data in the form of real-time data to be processed.

```python
# import dependencies
import os
from pandas import DataFrame, read_csv
```

```python
files = os.listdir("./data/")
for index, file in enumerate(files):
    temp_df = DataFrame(read_csv(f"./data/{file}"))
    temp_df.drop(columns=temp_df.columns[2:8], axis=0, inplace=True)
    temp_df.to_csv(f"./cropped_data/{index}.csv")
```
